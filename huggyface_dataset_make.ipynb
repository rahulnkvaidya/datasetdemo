## Initialize empty list and error list

### Subtask:
Create an empty list to store the loaded datasets from valid files and an empty list to store the filenames of files with errors.


**Reasoning**:
Initialize empty lists to store valid datasets and error files.



```python
valid_datasets = []
error_files = []
```

## Iterate and load

### Subtask:
Loop through each file in the list. Inside the loop, use a `try-except` block to load each JSON file individually into a dataset. If an error occurs during loading, add the filename to the error list. If loading is successful, append the loaded dataset to the list of valid datasets.


**Reasoning**:
Loop through each file, attempt to load it as a dataset, and handle errors by adding problematic filenames to a list.



```python
import os
from datasets import load_dataset

drive_dir = '/content/drive/MyDrive/aidataset/'

for filename in json_files:
    file_path = os.path.join(drive_dir, filename)
    try:
        dataset = load_dataset('json', data_files={'train': file_path}, split='train')
        valid_datasets.append(dataset)
    except Exception as e:
        print(f"Error loading file {filename}: {e}")
        error_files.append(filename)

print(f"Successfully loaded {len(valid_datasets)} datasets.")
print(f"Failed to load {len(error_files)} files: {error_files}")
```

## Concatenate datasets

### Subtask:
Concatenate all the datasets in the list of valid datasets into a single dataset.


**Reasoning**:
Import the `concatenate_datasets` function from the `datasets` library and use it to combine the valid datasets.



```python
from datasets import concatenate_datasets

concatenated_dataset = concatenate_datasets(valid_datasets)
print(concatenated_dataset)
```

## Save error filenames

### Subtask:
Save the list of error filenames to a text file in Google Drive.


**Reasoning**:
Define the path for the error file and write the filenames from the error_files list to it, with each filename on a new line.



```python
import os

# Define the full path for the error file in your Google Drive
error_file_path = '/content/drive/MyDrive/aidataset/error_files.txt'

# Open the specified file in write mode ('w')
with open(error_file_path, 'w') as f:
    # Iterate through the error_files list and write each filename to the file
    for filename in error_files:
        f.write(filename + '\n')

print(f"Error filenames saved to {error_file_path}")
```

## Save concatenated dataset to parquet

### Subtask:
Save the concatenated dataset to a Parquet file in Google Drive.


**Reasoning**:
Save the concatenated dataset to a Parquet file in Google Drive and print a confirmation message.



```python
import os

# Define the full path where the Parquet file should be saved in your Google Drive.
parquet_file_path = '/content/drive/MyDrive/aidataset/train.parquet'

# Use the to_parquet() method of the concatenated_dataset object to save the dataset.
concatenated_dataset.to_parquet(parquet_file_path)

# Print a confirmation message.
print(f"Concatenated dataset saved to Parquet file at: {parquet_file_path}")
```

## Upload to hugging face hub

### Subtask:
Upload the concatenated dataset (from the Parquet file or directly from the dataset object) to the Hugging Face Hub.


**Reasoning**:
Upload the concatenated dataset to the Hugging Face Hub using the `push_to_hub` method.
