```python
from datasets import Dataset

data = {
    "text": [
        "भारत की राजधानी दिल्ली है।",
        "मैं एक AI असिस्टेंट हूँ।",
        "आज मौसम बहुत अच्छा है।"
    ]
}

# Python डिक्शनरी से Hugging Face Dataset ऑब्जेक्ट बनाएँ।
dataset = Dataset.from_dict(data)

print(dataset)
```

```python
# 'train.parquet' नाम से Parquet फाइल बनाएँ।
dataset.to_parquet("train.parquet")
```

```python
from huggingface_hub import login

# Hugging Face में लॉगिन करें
login()  # यह आपको एक टोकन डालने के लिए प्रॉम्प्ट करेगा।

# डेटासेट को Hugging Face Hub पर अपलोड करें।
dataset.push_to_hub("rahulnkvaidya/demo") # Replace with your username and dataset name
```
